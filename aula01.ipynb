##Análise Assistida por IA


### Solicitar à IA:

1. Classificação dos dados;

2. Identificação de possíveis anomalias;

3. Sugestões de risco.

```python
import json

# 1. Carregando e Limpando os Dados do JSON
json_data = {
    "telemetria": {
        "temperatura": {"interna": "22°C", "externa": "-55°C"},
        "integridadeEstrutural": 1,
        "niveisEnergia": "85%",
        "pressaoTanques": "50 psi",
        "statusModulos": {
            "comunicacao": "Ativo", "navegacao": "Ativo", "controleVida": "Ativo"
        }
    }
}

# Função auxiliar para extrair apenas os números
def limpar_valor(texto):
    if isinstance(texto, str):
        # Remove símbolos e converte para float
        apenas_num = "".join(c for c in texto if c.isdigit() or c in '-.')
        return float(apenas_num)
    return texto

# Extraindo dados limpos
t_interna = limpar_valor(json_data['telemetria']['temperatura']['interna'])
t_externa = limpar_valor(json_data['telemetria']['temperatura']['externa'])
energia = limpar_valor(json_data['telemetria']['niveisEnergia'])
pressao = limpar_valor(json_data['telemetria']['pressaoTanques'])
integridade = json_data['telemetria']['integridadeEstrutural']

# --- ANÁLISE ASSISTIDA POR IA ---

print("--- RELATÓRIO DE ANÁLISE DE SISTEMAS ---")

# A. CLASSIFICAÇÃO DOS DADOS
print(f"\n[CLASSIFICAÇÃO]:")
status_energia = "CRÍTICO" if energia < 20 else "NOMINAL" if energia > 70 else "ALERTA"
print(f"- Energia: {energia}% ({status_energia})")
print(f"- Ambiente: Térmico Estável (Delta T: {abs(t_interna - t_externa)}°C)")

# B. IDENTIFICAÇÃO DE ANOMALIAS
anomalias = []
if integridade < 1:
    anomalias.append("Falha na integridade da carcaça")
if pressao > 100 or pressao < 10:
    anomalias.append("Pressão dos tanques fora da zona de segurança")
if t_interna > 30 or t_interna < 15:
    anomalias.append("Falha no controle térmico interno")

print(f"\n[ANOMALIAS]:")
if not anomalias:
    print("- Nenhuma anomalia detectada nos sistemas principais.")
else:
    for a in anomalias: print(f"- ATENÇÃO: {a}")

# C. SUGESTÕES DE RISCO
print(f"\n[MATRIZ DE RISCO]:")
if status_energia == "CRÍTICO" or integridade < 1:
    print("RISCO: ALTO - Abortar missão imediatamente.")
elif t_externa < -100:
    print("RISCO: MÉDIO - Monitorar congelamento de sensores externos.")
else:
    print("RISCO: BAIXO - Condições favoráveis para operação.")
```

A vontade de explorar o espaço sempre foi um grande sonho da humanidade. Mas, quando olhamos para as estrelas, não podemos esquecer de quem vive aqui embaixo. Buscar o desconhecido deixa de ser apenas uma aventura científica e vira uma questão de escolha: o que é mais importante agora? Muita gente acha que a ética (o que é certo ou errado) é algo que cada um decide por si, mas quando falamos de grandes empresas e foguetes, a responsabilidade tem que ser de todos.

Em primeiro lugar, existe uma briga entre o desejo de criar tecnologias incríveis e a necessidade de ajudar as pessoas. É estranho ver bilhões de reais sendo gastos em foguetes enquanto problemas básicos, como a fome ou a falta de saneamento, continuam sem solução. É por isso que existem tantos protestos e reuniões contra esses projetos. Se os cientistas e os donos dessas empresas pensarem apenas no sucesso de chegar ao espaço e esquecerem das pessoas mais pobres ou da saúde do nosso planeta, a ciência perde o seu valor de ajudar a humanidade e vira apenas vaidade.

Além disso, hoje em dia as empresas falam muito sobre "sustentabilidade". Isso significa que elas precisam criar tecnologias que não destruam o meio ambiente. Para que o nosso planeta não fique em uma situação pior do que já está, as empresas que fazem foguetes precisam diminuir o lixo que jogam no espaço e o quanto poluem o ar. A exploração do universo não pode ser feita de qualquer jeito; ela deve servir para ajudar a Terra também, como no caso dos satélites que avisam sobre tempestades ou incêndios. Não adianta nada conquistar outros planetas se a gente estragar a nossa única casa.

Por tudo isso, precisamos de regras para que o progresso da ciência caminhe junto com o cuidado com as pessoas e a natureza. Os governos deveriam criar leis que obriguem as empresas espaciais a usar parte do seu lucro para resolver problemas aqui na Terra. Também é importante que, desde a escola, os futuros engenheiros e cientistas aprendam a colocar a vida humana sempre em primeiro lugar. Só assim a tecnologia deixará de ser um peso para o mundo e será, de verdade, uma ajuda para o nosso futuro.
